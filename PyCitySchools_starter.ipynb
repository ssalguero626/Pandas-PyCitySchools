```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
schools_df = "Resources/schools_complete.csv"
students_df = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools_df = pd.read_csv(schools_df)
students_df = pd.read_csv(students_df)


# Combine the data into a single dataset.
combined_school_data = pd.merge(schools_df, students_df, how="left", on=["school_name", "school_name"])
combined_school_data.head()                       
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total number of Schools
total_schools = schools_df['school_name'].count()

# Total number of Students
total_students = students_df['student_name'].count()

# Total Budget
total_budget = schools_df['budget'].sum()

# Average math score
avg_math = students_df['math_score'].mean()

# Average reading score
avg_reading = students_df['reading_score'].mean()

# Percentage of passing math score and count
pass_math = students_df.loc[students_df['math_score'] >=70]
pass_math_count = pass_math['student_name'].count()

# Percentage of passing reading score and count
pass_reading = students_df.loc[students_df['reading_score'] >=70]
pass_reading_count = pass_reading['student_name'].count()

# Percentage of students who passed math and reading (% Overall Passing)
overall_pass = ((pass_math_count + pass_reading_count)/(2*total_students))


# Create DataFrame to hold results
district_summary = pd.DataFrame({"Total Schools": total_schools,
                                 "Total Students": [total_students],
                                 "Total Budget": [total_budget],
                                 "Average Math Score":[avg_math],
                                 "Average Reading Score": [avg_reading],
                                 "% Passing Math": [(pass_math_count/total_students)*100],
                                 "% Passing Reading": [(pass_reading_count/total_students)*100],
                                 "% Overall Passing Rate": [overall_pass*100]})
# Give DataFrame cleaner formatting
district_summary["Total Students"] = district_summary["Total Students"].astype(float).map("{:,.0f}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].astype(float).map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].astype(float).map("{:.1f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].astype(float).map("{:.1f}".format)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Group by Schools
school_names = combined_school_data.set_index("school_name").groupby(["school_name"])

# Group by Type
school_type = schools_df.set_index("school_name")["type"]

# Total Students
students_per_school = school_names['Student ID'].count()

# Total School Budget
school_budget = schools_df.set_index("school_name")["budget"]

# Per Student Budget
budget_per_student = schools_df.set_index("school_name")["budget"]/schools_df.set_index("school_name")["size"]

# Average Math score
average_math_score = school_names["math_score"].mean()

# Average Reading Score

average_reading_score = school_names["reading_score"].mean()

# % Percent Passing Math

passing_math_percent = combined_school_data[combined_school_data["math_score"] >= 70].groupby("school_name")["Student ID"].count()/students_per_school

# % Percent Passing Reading

passing_read_percent = combined_school_data[combined_school_data["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/students_per_school

# Overall Pass Rate

overall_pass_rate = (passing_math_percent + passing_read_percent)/2

# Create DataFrame to hold results

summary_by_school = pd.DataFrame({ "School Type": school_type,
                                   "Total Students": students_per_school,
                                   "Total School Budget": school_budget,
                                   "Per Student Budget": budget_per_student,
                                   "Average Math Score": average_math_score,
                                   "Average Reading Score": average_reading_score,
                                   "% Passing Math": passing_math_percent,
                                   "% Passing Reading": passing_read_percent,
                                   "Overall Passing Rate": overall_pass_rate})



summary_by_school.style.format({'Total Students': '{:,}',
                                'Total School Budget': "${:,}",
                                'Per Student Budget': "${:.0f}",
                                'Average Math Score': "{:.1f}",
                                'Average Reading Score': "{:.1f}"
                               
                                
                               })




```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_schools = summary_by_school.sort_values("Overall Passing Rate", ascending = False)
top_schools.head()
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
bottom_schools = summary_by_school.sort_values("Overall Passing Rate")
bottom_schools.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create spending bins

spending_bins = [0, 584, 629, 645, 675 ]
group_names = ['< $585', '$585 - $629', '$630 -$644', '$645-$675']

combined_school_data["spending_bins"] = pd.cut(combined_school_data["budget"]/combined_school_data["size"], spending_bins, labels = group_names)

# Group by spending
by_spending = combined_school_data.groupby('spending_bins')

# Calculations

avg_math = by_spending['math_score'].mean()
avg_read = by_spending['reading_score'].mean()
pass_math = combined_school_data[combined_school_data['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
pass_read = combined_school_data[combined_school_data['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()
overall = combined_school_data[(combined_school_data['reading_score'] >= 70) & (combined_school_data['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/by_spending['Student ID'].count()

# Create DataFrame

scores_by_spend = pd.DataFrame({
                    "Average Math Score": avg_math,
                    "Average Reading Score": avg_read,
                    '% Passing Math': pass_math,
                    '% Passing Reading': pass_read,
                    "Overall Passing Rate": overall
})

#Reorder columns

scores_by_spend = scores_by_spend[["Average Math Score",
                                  "Average Reading Score",
                                  '% Passing Math',
                                  '% Passing Reading',
                                  "Overall Passing Rate"]]

scores_by_spend.index.name = "Per Student Budget"

# Format for cleaner output
scores_by_spend.style.format({'Average Math Score': '{:.1f}',
                              'Average Reading Score': '{:.1f}',
                              '% Passing Math': '{:.1%}',
                              '% Passing Reading': '{:.1%}',
                              'Overall Passing Rate': '{:.1%}'
                              
                              
                             
                             })

```
